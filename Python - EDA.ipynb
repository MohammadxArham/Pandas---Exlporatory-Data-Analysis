```python
 # top 10 highest populated countries.
df.sort_values(by = "World Population Percentage", ascending = False).head(10)
```

```python
df.fillna('')
```

```python
 # avg. population per continent in order.
df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
# Every row with Oceania in its Continent column.
df[df['Continent'].str.contains('Oceania')] 
```

```python
# avg. population per continent per year's census in order.
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2   
```

```python
# Basic plotting of the derived table in a simple line graph.
df2.plot()
```

```python
# Exchanging the position of columns for better plotting.
df3 = df2.transpose()
df3
```

```python
# Increasing the size of the graph for better view.
df3.plot(figsize = (15,5))
```
